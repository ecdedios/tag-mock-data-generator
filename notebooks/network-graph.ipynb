```python
import pandas as pd
df = pd.read_csv('../data/out/data.csv')
```

```python
import networkx as nx
G = nx.Graph()
```

```python
G = nx.from_pandas_edgelist(df, 'PARENT', 'CHILD')
```

```python
from matplotlib.pyplot import figure
```

```python
figure(figsize=(10, 8))
nx.draw(G)
```

```python
leaderboard = {}
for x in G.nodes:
 leaderboard[x] = len(G[x])
s = pd.Series(leaderboard, name='connections')
df2 = s.to_frame().sort_values('connections', ascending=False)
```
